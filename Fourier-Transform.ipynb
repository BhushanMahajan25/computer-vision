```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_stripes = cv2.imread(r'CVND_Exercises-master\1_2_Convolutional_Filters_Edge_Detection\images\stripes.jpg')
image_stripes = cv2.cvtColor(image_stripes, cv2.COLOR_BGR2RGB)

image_solid = cv2.imread(r'CVND_Exercises-master\1_2_Convolutional_Filters_Edge_Detection\images\pink_solid.jpg')
image_solid = cv2.cvtColor(image_solid, cv2.COLOR_BGR2RGB)

f, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.imshow(image_stripes)
ax2.imshow(image_solid)
print('f',f)
```

```python
gray_stripes = cv2.cvtColor(image_stripes, cv2.COLOR_RGB2GRAY)
gray_solid = cv2.cvtColor(image_solid, cv2.COLOR_RGB2GRAY)

f, (ax3,ax4) = plt.subplots(1,2,figsize=(10,5))
ax3.imshow(gray_stripes, cmap='gray')
ax4.imshow(gray_solid, cmap='gray')
```

```python
norm_stripes = gray_stripes/255.0
norm_solid = gray_solid/255.0

f, (ax5,ax6) = plt.subplots(1,2,figsize=(10,5))
ax5.imshow(norm_stripes)
ax6.imshow(norm_solid)
```

```python
def ft_image(norm_image):
    f = np.fft.fft2(norm_image)
    f_shift = np.fft.fftshift(f)
    freq = 20*np.log(np.abs(f_shift))
    return freq
```

```python
f_stripes = ft_image(norm_stripes)
f_solid = ft_image(norm_solid)
```

```python
f, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10))
ax1.imshow(image_stripes)
ax1.set_title('original image')
ax2.imshow(f_stripes, cmap='gray')
ax2.set_title('FT image')

ax3.imshow(image_solid)
ax3.set_title('original image')
ax4.imshow(f_solid, cmap='gray')
ax4.set_title('FT image')
```

```python
image = cv2.imread(r'CVND_Exercises-master\1_2_Convolutional_Filters_Edge_Detection\images\birds.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image_solid, cv2.COLOR_RGB2GRAY)
norm_image = gray_image/255.0

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.imshow(image)
ax1.set_title('original image')
ax2.imshow(gray_image, cmap='gray')
ax2.set_title('grayscale image')
ax3.imshow(norm_image)
ax3.set_title('normalized image')
```

```python
f_image = ft_image(norm_image)
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(image)
ax2.imshow(f_image, cmap='gray')
```
